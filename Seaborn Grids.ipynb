## Seaborn Grids

Grids are general types of plots that allow one to map plot types to rows and columns of a grid. This helps to create similar plots separated by features.

- Pairgrid
- Facetgrid
- Jointgrid

```python
# Imports
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.shape
```

## Pairgrid

**Pairgrid** is a subplot grid for plotting pairwise relationships in a dataset.

```python
sns.PairGrid(iris)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
plt.show()
```

## Pairplot (simpler version of Pairgrid)

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
plt.show()
```

## Facet grid

**Facet Grid** is the general way to create grids of plots based off of a feature

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

## Jointgrid

**Joint Grid** is the general version for jointplot() type grids

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
plt.show()
```
